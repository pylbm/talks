The lattice Boltzmann method needs a stencil of velocities. Each distribution of particles $f$ is associated to a velocity $c\in\lambda\mathbb{Z}^d$ where $\lambda=dx/dt$ is the scheme velocity and $d$ the spatial dimension.

The velocities are numbered with the natural convention of the LBM. 

```python
import pyLBM
import numpy as np
%config InlineBackend.figure_format = 'retina'
```

<div style="color: navy; background-color: skyblue; border: 2px solid navy; padding: 5px;">
  <p style="text-align: center;font-size: 150%">In dimension 1</p>
</div>

In dimension 1, the velocities are numbered as:
* $c_0 = 0$,
* $c_1 = 1$, $c_2 = -1$,
* $c_3 = 2$, $c_4 = -2$,
* $c_5 = 3$, $c_6 = -3$, $\ldots$

The stencil can be created and visualized as in the following examples.

> Note that the dimension has to be given to the constructor with the keyword `dim`. This keyword is optional and is not needed if the geometry is known (and in particular if the key `box` can be used to compute the spatial dimension)

```python
s = pyLBM.Stencil({
    'dim': 1,
    'schemes':[
        {'velocities': list(range(9))}, 
    ],
})
print(s)
s.visualize()
```

It is possible to define more than one stencil in order to allowed vectorial schemes. The value associated to the key `schemes` is a list that can contain more than one scheme.

```python
s = pyLBM.Stencil({
    'dim': 1,
    'schemes':[
        {'velocities': [1,2]}, 
        {'velocities': [0,1,2]}, 
    ],
})
print(s)
s.visualize()
```

<div style="color: navy; background-color: skyblue; border: 2px solid navy; padding: 5px;">
  <p style="text-align: center;font-size: 150%">In dimension 2</p>
</div>

In dimension 2, the velocities are numbered as:
* $c_0 = (0,0)$,
* $c_1 = (1,0)$, $c_2 = (0,1)$, $c_3 = (-1,0)$, $c_4 = (0,-1)$,
* $c_5 = (1,1)$, $c_6 = (-1,1)$, $c_7 = (-1,-1)$, $c_8=(1,-1)$, $\ldots$

```python
s = pyLBM.Stencil({
    'dim': 2,
    'schemes':[
        {'velocities':range(9)},
        {'velocities':range(13)},
        {'velocities':range(25)},
    ],
})
print(s)
s.visualize()
```

<div style="color: navy; background-color: skyblue; border: 2px solid navy; padding: 5px;">

Then the classical stencils in dimension 3:

```python
s = pyLBM.Stencil({
    'dim': 3,
    'schemes':[
        {'velocities':range(7)},
        {'velocities':range(19)},
        {'velocities':range(27)},
    ],
})
print(s)
s.visualize()
```
